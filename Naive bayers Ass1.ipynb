Q1. What is Bayes' theorem?

Bayes’ theorem describes the probability of occurrence of an event related to any condition. It is also considered for the case of conditional probability. Bayes theorem is also known as the formula for the probability of “causes”. For example: if we have to calculate the probability of taking a blue ball from the second bag out of three different bags of balls, where each bag contains three different colour balls viz. red, blue, black. In this case, the probability of occurrence of an event is calculated depending on other conditions is known as conditional probability. In this article, let us discuss the statement and proof for Bayes theorem, its derivation, formula, and many solved examples.

Q2. What is the formula for Bayes' theorem?

P(A|B)=P(B|A)*P(A)/P(B)

Here, P(A) = how likely A happens(Prior knowledge)- The probability of a hypothesis is true before any evidence is present.

P(B) = how likely B happens(Marginalization)- The probability of observing the evidence.

P(A|B) = how likely A happens given that B has happened(Posterior)-The probability of a hypothesis is true given the evidence.

P(B|A) = how likely B happens given that A has happened(Likelihood)- The probability of seeing the evidence if the hypothesis is true.

Q4. What is the relationship between Bayes' theorem and conditional probability?

Conditional Probability:-
1.Conditional Probability is the probability of occurrence of a certain event, say A based on some other event whether B is true or not
2.The equation of conditional probability is:P(A|B)=P(A∩B)P(B)
3.It is used to compute the conditional probability and the events A and B are relatively simple
4.It is used for relatively simple problems.
Bayes Theorem:-
1.Bayes Theorem includes two conditional probabilities for the events, say A and B
2.The equation of Bayes Theorem is: P(A|B)=P(B|A)×P(A)P(B)
3.It is used in Bayesian inference and in models where we are interested in the distribution up to a normalizing factor P(B)
4.It gives a structured formula for solving more complex problems.

Q2. What is the objective function of a linear SVM?

The objective function is the geometric margin of the hyperplane (w, b). The constraints represent the fact that the objective function is the minimum of the set of geometric margins of the hyperplane (w, b) w.r.t. all the training examples

Q5. Illustrate with examples and graphs of Hyperplane, Marginal plane, Soft margin and Hard margin in
SVM?

Support Vector Machines (SVM) are powerful machine learning models used for classification and regression tasks. They work by finding the optimal hyperplane that separates data into different classes. In SVM, you have the concepts of hyperplane, marginal plane, soft margin, and hard margin, each with its own significance. I'll provide examples and explanations for each of these concepts:

1.Hyperplane:
A hyperplane is a linear decision boundary that separates data into different classes. In a binary classification problem, it can be represented as follows:

For a 2D space (two features), the equation of the hyperplane is: w1*x1 + w2*x2 + b = 0, where (w1, w2) are the weights, (x1, x2) are the input features, and b is the bias term.

Example: Imagine you have a dataset with two classes of points, red and blue, and you want to separate them. The hyperplane can be a line that best divides the two classes

2.Marginal Plane:
The marginal plane, also known as the support vector plane, is the hyperplane that is closest to the data points from both classes. These data points are called support vectors. The distance from this plane to the nearest support vector is known as the margin. The SVM algorithm tries to maximize this margin while minimizing classification errors.

Example: In the same dataset with red and blue points, the marginal plane is the one that is equidistant from the nearest red and blue points, maximizing the margin between them.

3.Hard Margin:
In a hard margin SVM, the goal is to find a hyperplane that perfectly separates the two classes without any misclassifications. This approach is suitable when the data is linearly separable, and it enforces a strict boundary.

Example: If your data is linearly separable, an SVM with a hard margin will find the hyperplane without allowing any data points to cross the margin, as shown below.

4.Soft Margin:
In real-world scenarios, data is often not perfectly separable, and there might be some noise or outliers. Soft margin SVMs allow for a certain level of misclassification. The margin is not required to be perfectly wide, and some data points are allowed to fall within the margin or even on the wrong side of the hyperplane. This is achieved by introducing a regularization parameter (C) that controls the trade-off between maximizing the margin and minimizing classification errors.

Example: For a dataset with some overlapping points or outliers, a soft margin SVM finds a hyperplane that allows for a certain number of misclassifications while still maximizing the margin.

The choice between a hard margin and soft margin SVM depends on the nature of the data. Hard margin SVMs are suitable for perfectly separable data, while soft margin SVMs are more robust when the data is not linearly separable or contains noise. The regularization parameter C in a soft margin SVM allows you to control the trade-off between a wider margin and a lower number of misclassifications.

Q6. SVM Implementation through Iris dataset.
~ Load the iris dataset from the scikit-learn library and split it into a training set and a testing setl
~ Train a linear SVM classifier on the training set and predict the labels for the testing setl
~ Compute the accuracy of the model on the testing setl
~ Plot the decision boundaries of the trained model using two of the featuresl
~ Try different values of the regularisation parameter C and see how it affects the performance of

```python
from  sklearn.datasets import load_iris
import pandas as pd
import numpy as np
```

```python
dataset=load_iris()
```

```python
df=pd.DataFrame(data=dataset.data,columns=dataset.feature_names)
```

```python
df.head()
```

```python
df.columns
```

```python
##dependent and independent features
X=df
y=dataset.target
```

```python
y
```

```python
X.shape,y.shape
```

```python
from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test=train_test_split(X,y,random_state=42,test_size=0.33)
```

```python
X_test.shape,y_test.shape
```

```python
from sklearn.svm import SVC
```

```python
svc=SVC(kernel='linear')
```

```python
svc.fit(X_train,y_train)
```

```python
y_pred=svc.predict(X_test)
```

```python
from sklearn.metrics import accuracy_score,confusion_matrix,classification_report
print(confusion_matrix(y_test,y_pred))
print(accuracy_score(y_test,y_pred))
print(classification_report(y_pred,y_pred))
```

```python
import seaborn as sns
sns.scatterplot(x=X_train['sepal length (cm)'],y=X_train["sepal width (cm)"],hue=y_train)
```

```python
from sklearn.model_selection import GridSearchCV
param_gird={'kernel':['linear','poly', 'rbf','sigmoid'],
            'C':[1,2,3,0.1,4,5],
            'gamma':['scale','auto']
           }
```

```python
grid=GridSearchCV(svc,param_grid=param_gird,cv=5,scoring='accuracy')
```

```python
grid.fit(X_train,y_train)
```

```python
grid.best_params_
```

```python
y_pred=grid.predict(X_test)
```

```python
from sklearn.metrics import accuracy_score,confusion_matrix,classification_report
print(confusion_matrix(y_test,y_pred))
print(accuracy_score(y_test,y_pred))
print(classification_report(y_pred,y_pred))
```
